```python
#clean data
def CleanResult(filepath):
    import pandas as pd
    exeldata = pd.read_excel(filepath)
    data = pd.DataFrame(exeldata)
    resultat_columns = [_ for _ in data.columns if "Résultat" in _]
    column_names = ["Code_Module","V_count_Session1","NV_count_Session1","AC_count_Session1","NRA_count_Session1","RAT_count_Session1","V_count_Session2","NV_count_Session2","AC_count_Session2","NRA_count_Session2","RAT_count_Session2","Date"]

    df = pd.DataFrame(columns = column_names)
    for i,element in enumerate(resultat_columns):
        actual_module_name = element.partition(':')[2].strip()
        file_name = os.path.basename(filepath)
        file_name_without_extention = os.path.splitext(file_name)[0]
        Date = file_name_without_extention.partition('-')[2].strip()

        
        df.at[i,"Code_Module"]= actual_module_name
        df.at[i,"Date"]= Date
        
        
        #V_count_Session1 = exeldata.apply(lambda x : True if exeldata['Résultat : '+actual_module_name] == "V" & exeldata['Session : '+actual_module_name] == 1 else False, axis = 1)
        V_count_Session1 = len(exeldata.loc[(exeldata['Résultat : '+actual_module_name] == 'V') & (exeldata['Session : '+actual_module_name] == 1)])
        NV_count_Session1 = len(exeldata.loc[(exeldata['Résultat : '+actual_module_name] == 'NV') & (exeldata['Session : '+actual_module_name] == 1)])
        AC_count_Session1 = len(exeldata.loc[(exeldata['Résultat : '+actual_module_name] == 'AC') & (exeldata['Session : '+actual_module_name] == 1)])
        NRA_count_Session1 = len(exeldata.loc[(exeldata['Résultat : '+actual_module_name] == 'NV') & (exeldata['Session : '+actual_module_name] == 1)])
        RAT_count_Session1 = len(exeldata.loc[(exeldata['Résultat : '+actual_module_name] == 'NV') & (exeldata['Session : '+actual_module_name] == 1)])
        
        V_count_Session2 = len(exeldata.loc[(exeldata['Résultat : '+actual_module_name] == 'V') & (exeldata['Session : '+actual_module_name] == 2)])
        NV_count_Session2 = len(exeldata.loc[(exeldata['Résultat : '+actual_module_name] == 'NV') & (exeldata['Session : '+actual_module_name] == 2)])
        AC_count_Session2 = len(exeldata.loc[(exeldata['Résultat : '+actual_module_name] == 'AC') & (exeldata['Session : '+actual_module_name] == 2)])
        NRA_count_Session2 = len(exeldata.loc[(exeldata['Résultat : '+actual_module_name] == 'NV') & (exeldata['Session : '+actual_module_name] == 2)])
        RAT_count_Session2 = len(exeldata.loc[(exeldata['Résultat : '+actual_module_name] == 'NV') & (exeldata['Session : '+actual_module_name] == 2)])
        
        df.at[i,"V_count_Session1"] = V_count_Session1
        df.at[i,"NV_count_Session1"] = NV_count_Session1
        df.at[i,"AC_count_Session1"] = AC_count_Session1
        df.at[i,"NRA_count_Session1"] = NRA_count_Session1
        df.at[i,"RAT_count_Session1"] = RAT_count_Session1
        
        df.at[i,"V_count_Session2"] = V_count_Session2
        df.at[i,"NV_count_Session2"] = NV_count_Session2
        df.at[i,"AC_count_Session2"] = AC_count_Session2
        df.at[i,"NRA_count_Session2"] = NRA_count_Session2
        df.at[i,"RAT_count_Session2"] = RAT_count_Session2

    df=df.fillna(0)
    #save into exel
    writer = pd.ExcelWriter("./../Cleaned data/"+file_name_without_extention+'_counts.xlsx')
    df.to_excel(writer)
    writer.save()
    return df
```

```python
def get_mining_data():
    #get all file names if the directory
    cleaned_filenames = next(walk("./Taux d'obtenation de diplome/"), (None, None, []))[2]  # [] if no file
    
    inscrit2014 = pd.read_excel("./Data inscrit/inscrt2014.xlsx")
    inscrit2015 = pd.read_excel("./Data inscrit/inscrt2015.xlsx")
    Total_Inscrit = pd.concat([inscrit2015, inscrit2014], axis=0)
    
    Mining_Data = pd.DataFrame()
    vertical_stack = pd.DataFrame()
    for name in cleaned_filenames:
        
        tmp_data = pd.read_excel("./Taux d'obtenation de diplome/"+name)
        COD_ETU_columns = [_ for _ in tmp_data.columns if "COD_ETU" in _]
        
        Note_columns = [_ for _ in tmp_data.columns if "Note" in _]

        Résultat_columns = [_ for _ in tmp_data.columns if "Résultat" in _]

        Session_columns = [_ for _ in tmp_data.columns if "Session" in _]
    
        Année_columns = [_ for _ in tmp_data.columns if "Année" in _]
        for i in range(len(Résultat_columns)):
            Mining_Data['COD_ETU'] = tmp_data[COD_ETU_columns[0]]
            Mining_Data['Module'] = Année_columns[i].partition(':')[2].strip()
            Mining_Data['Note_Module'] = tmp_data[Note_columns[i]]
            Mining_Data['Résultat_Module'] = tmp_data[Résultat_columns[i]]
            Mining_Data['Session_Module'] = tmp_data[Session_columns[i]]
            Mining_Data['Année_Module'] = tmp_data[Année_columns[i]]
            vertical_stack = pd.concat([vertical_stack, Mining_Data], axis=0)
        
        dff = pd.merge(vertical_stack, Total_Inscrit,on='COD_ETU')

    return dff
```
